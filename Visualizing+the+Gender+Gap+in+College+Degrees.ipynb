```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv(r'C:/Users/Green Miracle/csv files/percent-bachelors-degrees-women-usa.csv')

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_light_gray = (171/255, 171/255, 171/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
ticks = [1970, 1980, 1990, 2000, 2010]

fig = plt.figure(figsize=(18, 30))

for sp in range(1, 17, 3):
    index = int(sp/3)
    
    ax = fig.add_subplot(6, 3, sp)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[index]], c=cb_orange, label='Men', linewidth=3)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_xticks(ticks)
    ax.set_yticks([0, 100])
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.axhline(50, c=cb_light_gray, alpha=0.3)
    
    if index == 0:
        ax.text(2005, 81, 'Women')
        ax.text(2007, 17, 'Men')
    elif index == 5:
        ax.text(2007, 86, 'Men')
        ax.text(2005, 10, 'Women')
        ax.tick_params(labelbottom=True)

    
for sp in range(2, 15, 3):
    index = int(sp/3)
    
    ax = fig.add_subplot(6, 3, sp)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[index]], c=cb_orange, label='Men', linewidth=3)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_xticks(ticks)
    ax.set_yticks([0, 100])
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.axhline(50, c=cb_light_gray, alpha=0.3)
    
    if index == 0:
        ax.text(2005, 74, 'Women')
        ax.text(2007, 23, 'Men')
    if index == 4:
        ax.tick_params(labelbottom=True)
    
for sp in range(3, 21, 3):
    index = int((sp-1)/3)
    
    ax = fig.add_subplot(6, 3, sp)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[index]], c=cb_orange, label='Men', linewidth=3)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
        
    ax.set_xlim(1968, 2011)
    ax.set_xticks(ticks)
    ax.set_yticks([0, 100])
    ax.set_ylim(0,100)
    ax.set_title(other_cats[index])
    ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
    ax.axhline(50, c=cb_light_gray, alpha=0.3)
    
    if index == 0:
        ax.text(2005, 88, 'Women')
        ax.text(2007, 9, 'Men')
    elif index == 5:
        ax.text(2007, 60, 'Men')
        ax.text(2005, 37, 'Women')
        ax.tick_params(labelbottom=True)
    

    
plt.savefig(r'C:/Users/Green Miracle/plots/gender_degrees.png')
plt.show()
```
